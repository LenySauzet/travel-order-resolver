```python
import dateparser
from datetime import datetime

expressions = [
    "mardi à 8h30",
    "17h",
    "ce soir",
    "dans la matinée",
    "le 22 avril",
    "demain matin",
    "samedi prochain à 15h",
]

for expr in expressions:
    parsed = dateparser.parse(
        expr,
        languages=['fr'],
        settings={
            'PREFER_DATES_FROM': 'future',
            'PREFER_DAY_OF_MONTH': 'first',
            'RETURN_AS_TIMEZONE_AWARE': False,
        }
    )
    if parsed:
        print(f"'{expr}' → {parsed.isoformat()}")
```
